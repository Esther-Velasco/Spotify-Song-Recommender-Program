```python
sp = spotipy.Spotify(auth_manager=SpotifyClientCredentials(client_id= config.client_id,
                                                           client_secret= config.client_secret))
```

## 2. Function to get playlist tracks

```python
def get_playlist_tracks(username, playlist_id):
    results = sp.user_playlist_tracks(username,playlist_id,market="GB")  
    tracks = results['items']  #tracks stores the first hundred songs of the playlist having a total of 250
    while results['next']:  #while there is a possibility to go further through playlist( more songs)
        results = sp.next(results)  #get the next page, store it in results
        tracks.extend(results['items']) #stores the list together in only one list(append would create list inside list)
    return tracks
```
